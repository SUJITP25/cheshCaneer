```python
from scipy.stats import chi2_contingency
import pandas as pd

chi2_test = []

for col in categorical_columns:
    p_value = chi2_contingency(pd.crosstab(df["case_status"], df[col]))[1]
    if p_value < 0.05:
        chi2_test.append("Reject Null Hypothesis")
    else:
        chi2_test.append("Fail to Reject Null Hypothesis")

# Correct DataFrame construction
result = pd.DataFrame({
    "Column": categorical_columns,
    "Hypothesis Result": chi2_test
})

result
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder, PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Remove duplicate
ordinal_columns = ["education_of_employee", "has_job_experience", "full_time_position"]
oh_columns = ["continent", "unit_of_wage", "region_of_employment"]
transform_columns = ["no_of_employees", "prevailing_wage"]

# Define transformers
oh_transformer = OneHotEncoder(handle_unknown='ignore')
ordinal_encoder = OrdinalEncoder()
power_transformer = PowerTransformer(method="yeo-johnson")

# Define ColumnTransformer
preprocessor = ColumnTransformer([
    ("onehot", oh_transformer, oh_columns),
    ("ordinal", ordinal_encoder, ordinal_columns),
    ("power", power_transformer, transform_columns)
])

# Example: Fit on DataFrame `df`
# (Assuming you already have `x` as a DataFrame of features)
x_transformed = preprocessor.fit_transform(df[oh_columns + ordinal_columns + transform_columns])
x_transformed
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay,precision_score,recall_score,fbeta_score,roc_auc_score,roc_curve,f1_score

from xgboost import XGBClassifier 
from catboost import CatBoostClassifier
```

```python
def evaluate_clf(true,predicted): 
    acc = accuracy_score(true,predicted)
    f1 =  f1_score(true,predicted)
    precision = precision_score(true,predicted)
    recall = recall_score(true,predicted)
    roc_auc = roc_auc_score(true,predicted)
    return acc,f1,precision,recall,roc_auc


models = {
    "Random Foerst" : RandomForestClassifier(), 
    "Decision Tree" : DecisionTreeClassifier(),
    "Gradient Boosting" : GradientBoostingClassifier(), 
    "LogisticRegression" : LogisticRegression(), 
    "XGBClassifier" : XGBClassifier(), 
    "CatBoosting Classifier" : CatBoostClassifier(verbose=False), 
    "Support Vector Classifier" : SVC(), 
    "AdaBoost Classifier" :AdaBoostClassifier()
}
```

```python
def evaluate_performance(x, y, models):
    from sklearn.model_selection import train_test_split

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

    results = []

    for i in range(len(models)):
        model = list(models.values())[i]
        name = list(models.keys())[i]

        model.fit(x_train, y_train)

        y_train_pred = model.predict(x_train)
        y_test_pred = model.predict(x_test)

        train_accuracy, train_f1, train_precision, train_recall, train_roc_auc = evaluate_clf(y_train, y_train_pred)
        test_accuracy, test_f1, test_precision, test_recall, test_roc_auc = evaluate_clf(y_test, y_test_pred)

        results.append({
            "Model": name,
            "Train Accuracy": train_accuracy,
            "Test Accuracy": test_accuracy,
            "Test F1": test_f1,
            "Test AUC": test_roc_auc
        })

    return pd.DataFrame(results)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define hyperparameters
xgboost_params = {
    "max_depth": range(3, 10, 2), 
    "min_child_weight": range(1, 6, 2)
}

rf_params = { 
    "max_depth": [10, 12, None, 15, 20], 
    "max_features": ["sqrt", "log2", None], 
    "n_estimators": [10, 50, 100, 200]
}

knn_params = {
    "algorithm": ["auto", "ball_tree", "kd_tree", "brute"], 
    "weights": ["uniform", "distance"], 
    "n_neighbors": [3, 4, 5, 7, 9]
}

# Combine models and params
randomcv_models = [
    ("XGBoost", XGBClassifier(use_label_encoder=False, eval_metric='logloss'), xgboost_params), 
    ("RandomForest", RandomForestClassifier(), rf_params), 
    ("KNN", KNeighborsClassifier(), knn_params)
]

# Dictionary to store best parameters
model_param = {}

# Run RandomizedSearchCV
for name, model, params in randomcv_models: 
    print(f"\n🔍 Running RandomizedSearchCV for {name}...")
    random = RandomizedSearchCV(
        estimator=model,
        param_distributions=params,
        n_iter=10,        # use 100 for real tuning
        cv=3,
        verbose=2,
        n_jobs=-1,
        random_state=42
    )
    random.fit(x_res, y_res)
    model_param[name] = random.best_params_

# Print best parameters
for model_name in model_param: 
    print(f"\n---------- Best Params for {model_name} ----------")
    print(model_param[model_name])
```
